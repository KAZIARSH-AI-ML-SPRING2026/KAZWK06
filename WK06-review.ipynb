```python
import matplotlib.pyplot as plt

from os import listdir
from PIL import Image as PImage

from image_utils import get_pixels

from Homework04_utils import Homework04Utils
```

```python
# List images

# Separate into train/test

all_clouds = listdir("./data/image/clouds")
all_flowers = listdir("./data/image/flowers")

print(len(all_clouds))
print(len(all_flowers))

train_clouds = all_clouds[:300]
train_flowers = all_flowers[:300]
```

```python
# iterate through all images
# get blue and/or green ratios

clouds_b_ratio = []
flower_b_ratio = []

clouds_g_ratio = []
flower_g_ratio = []

correct_cloud = 0
correct_flower = 0

def is_cloud(bratio):
  return bratio > 0.4

def is_flower1(b, g):
  return g > 0.4 and g > b

def is_flower(b, g):
  return g - b > 0.2


for f in train_clouds:
  img = PImage.open("./data/image/clouds/" + f)
  img = img.resize((64,64))
  pxs = get_pixels(img)
  blue_ratio = Homework04Utils.color_ratio(pxs, (0, 100, 200), 150)
  green_ratio = Homework04Utils.color_ratio(pxs, (0, 100, 20), 150)
  clouds_b_ratio.append(blue_ratio)
  clouds_g_ratio.append(green_ratio)
  if is_flower(blue_ratio, green_ratio) == False:
    correct_cloud += 1


for f in train_flowers:
  img = PImage.open("./data/image/flowers/" + f)
  img = img.resize((64,64))
  pxs = get_pixels(img)
  blue_ratio = Homework04Utils.color_ratio(pxs, (0, 100, 200), 150)
  green_ratio = Homework04Utils.color_ratio(pxs, (0, 100, 20), 150)
  flower_b_ratio.append(blue_ratio)
  flower_g_ratio.append(green_ratio)
  if is_flower(blue_ratio, green_ratio):
    correct_flower += 1

print(correct_cloud/len(train_clouds))
print(correct_flower/len(train_flowers))
```

```python
# plot blue and/or green ratios for both classes of images

plt.scatter(len(clouds_b_ratio) * [0], clouds_b_ratio, s=4, alpha=0.1, color="#0000ff")
plt.scatter(len(clouds_g_ratio) * [1], clouds_g_ratio, s=4, alpha=0.1, color="#00ff00")

plt.scatter(len(flower_b_ratio) * [3], flower_b_ratio, s=4, alpha=0.1, color="#0000ff")
plt.scatter(len(flower_g_ratio) * [4], flower_g_ratio, s=4, alpha=0.1, color="#00ff00")

plt.show()
```
